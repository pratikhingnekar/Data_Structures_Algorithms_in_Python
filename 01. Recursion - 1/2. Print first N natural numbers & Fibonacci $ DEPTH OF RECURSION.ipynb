```python
def Print1toN(n):
    if n ==0:
        return
    Print1toN(n-1)
    print(n)
    
n = int(input())
Print1toN(n)
```

```python
def PrintNto1(n):
    if n ==0:
        return
    print(n)
    PrintNto1(n-1)
    
n = int(input())
PrintNto1(n)
```

```python
def fib(n):
    if n==1 or n==2:
        return 1
    fib_n_1=fib(n-1)
    fib_n_2=fib(n-2)
    output=fib_n_1+fib_n_2
    return output
```

```python
n = 5
fib(n)
```

### FACTORIAL

```python
def fact(n):
    if n == 0:
        return 1
    return n*fact(n-1)

n = int(input())
fact(n)
```
